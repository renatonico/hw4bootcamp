# PyCity Schools Analysis

* Math scores for the district are, on average, 4 points below reading scores. This is a reflection of the large number of students in the bottom performing schools, where this difference is seen. The top performing schools have close to the same scores for math and reading.
* Within each school, both reading and math scores do not vary significantly from 9th to 12th grade, although there are significant differences between schools
* There seems to be a great divide within the larger schools. Analysing the results per student spend in math, average scores for the bottom performers are only about 7 points below the top performing schools. However, the percentage of students passing is about 28% lower, and the overall passing rate is 36% lower. This indicates there is a pool of very high scorers in schools with close to 55% overall passing rates.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
#school_data_complete_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#=======================================
# = DISTRICT-WIDE TOTALS AND AVERAGES  =
#=======================================

# From the original school_data_df dataframe:
numsch=school_data_df["school_name"].count()
numstu=school_data_df["size"].sum()
totbud=school_data_df["budget"].sum()

# From the original student_data_df dataframe:
avgmat=student_data_df["math_score"].mean()
avgrea=student_data_df["reading_score"].mean()
passma=len(student_data_df[student_data_df["math_score"]>=70])
passmp=passma/numstu
passre=len(student_data_df[student_data_df["reading_score"]>=70])
passrp=passre/numstu
totpass= student_data_df[(student_data_df["math_score"]>=70) & (student_data_df["reading_score"]>=70)].shape[0]
ovpass = totpass/numstu
```

```python
#================================================
#= CLEAN UP AND COMBINE VALUES INTO A DATAFRAME =
#================================================

# Rename columns and convert single series into a dataframe for display
District_Summary_df = pd.DataFrame({"Total Schools": [numsch],
                                "Total Students": [numstu],
                                "Total Budget":[totbud],
                                "Avg. Math Score": [avgmat],
                                "Avg. Reading Score":[avgrea],
                                "% Passed Math": [passmp],
                                "% Passed Reading": [passrp],
                                "% Overall Passing Rate": [ovpass]})

# Format values 
District_Summary_df["Total Budget"] = District_Summary_df["Total Budget"].map("${:,}".format)
District_Summary_df["Avg. Math Score"] = District_Summary_df["Avg. Math Score"].map("{:.2f}".format)
District_Summary_df["Avg. Reading Score"] = District_Summary_df["Avg. Reading Score"].map("{:.2f}".format)
District_Summary_df["% Passed Math"] = District_Summary_df["% Passed Math"].map("{:.2%}".format)
District_Summary_df["% Passed Reading"] = District_Summary_df["% Passed Reading"].map("{:.2%}".format)
District_Summary_df["% Overall Passing Rate"] = District_Summary_df["% Overall Passing Rate"].map("{:.2%}".format)

# Display dataframe
District_Summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#============================================================
#= CALCULATE THE DATA TO ADD TO THE SCHOOL_DATA DATAFRAME =
#============================================================

# Since the original school_data_df already has the first 4 items tabled, it's easy to add the "per student budget"
school_data_df["per_student"]=school_data_df["budget"]/school_data_df["size"]

# For the averages for each school, groupby the merged df by school name and put the averages in dataframes to merge later
School_Summary_df = school_data_complete_df.groupby(["school_name"])
mathavg_df=pd.DataFrame(School_Summary_df["math_score"].mean())
readavg_df=pd.DataFrame(School_Summary_df["reading_score"].mean())

# For the percentages of passing grades, create dataframes with the number of passing,
# then merge with school_data_df to calculate the percent based on total students in each school
passmath_df=pd.DataFrame(school_data_complete_df[school_data_complete_df["math_score"] >=70].groupby("school_name")["grade"].count())
passmath_df=passmath_df.rename(columns={"grade":"pass_math"})
passread_df=pd.DataFrame(school_data_complete_df[school_data_complete_df["reading_score"] >=70].groupby("school_name")["grade"].count())
passread_df=passread_df.rename(columns={"grade":"pass_read"})
passover_df=pd.DataFrame(school_data_complete_df[(school_data_complete_df["math_score"] >=70) & (school_data_complete_df["reading_score"] >=70)].groupby("school_name")["grade"].count())
passover_df=passover_df.rename(columns={"grade":"pass_over"})

#======================================
#= MERGE THE CALCULATED COLUMN DATA =
#======================================

# Merge average scores and passing numbers with school_data_df
school_data_ma_df = pd.merge(school_data_df, mathavg_df, on="school_name")
school_data_mara_df = pd.merge(school_data_ma_df, readavg_df, on="school_name")
school_data_marapm_df = pd.merge(school_data_mara_df,passmath_df, on="school_name")
school_data_marapmpr_df = pd.merge(school_data_marapm_df,passread_df, on="school_name")
school_data_stats_df = pd.merge(school_data_marapmpr_df,passover_df, on="school_name")

# Calculate %passing and add to the school_data_stats_df
school_data_stats_df["%pass_math"]=school_data_stats_df["pass_math"]/school_data_stats_df["size"]
school_data_stats_df["%pass_read"]=school_data_stats_df["pass_read"]/school_data_stats_df["size"]
school_data_stats_df["%overall"]=school_data_stats_df["pass_over"]/school_data_stats_df["size"]

#================
#=  CLEAN UP    =
#================

# Delete number of students used to calculate the percentages
del school_data_stats_df["pass_math"]
del school_data_stats_df["pass_read"]
del school_data_stats_df["pass_over"]
#delete the unnecessary "school id" column
del school_data_stats_df["School ID"]

#Rename columns
school_data_stats_df = school_data_stats_df.rename(columns={"school_name":"School Name",
                                                            "type":"School Type","size":"Total Students",
                                                            "budget":"School Budget",
                                                            "per_student":"Per Student Budget",
                                                            "math_score":"Average Math Score",
                                                            "reading_score":"Average Reading Score",
                                                            "%pass_math":"% Passing Math",
                                                            "%pass_read":"% Passing Reading",
                                                            "%overall":"% Overall Passing Rate"})

# Save 3 copies of the final dataframe without the number formatting 
# to be able to perform other operations on the values during the binning reports
school_data_stats_spend_df = school_data_stats_df.copy()
school_data_stats_size_df = school_data_stats_df.copy()
school_data_stats_type_df = school_data_stats_df.copy()

# Format numbers
school_data_stats_df["School Budget"] = school_data_stats_df["School Budget"].map("${:,.2f}".format)
school_data_stats_df["Per Student Budget"] = school_data_stats_df["Per Student Budget"].map("${:,.2f}".format)
school_data_stats_df["Average Math Score"] = school_data_stats_df["Average Math Score"].map("{:.2f}".format)
school_data_stats_df["Average Reading Score"] = school_data_stats_df["Average Reading Score"].map("{:.2f}".format)
school_data_stats_df["% Passing Math"] = school_data_stats_df["% Passing Math"].map("{:.2%}".format)
school_data_stats_df["% Passing Reading"] = school_data_stats_df["% Passing Reading"].map("{:.2%}".format)
school_data_stats_df["% Overall Passing Rate"] = school_data_stats_df["% Overall Passing Rate"].map("{:.2%}".format)

# Index by school
school_data_stats_df = school_data_stats_df.set_index("School Name")

school_data_stats_df
```
